```python
dropdata=data.iloc[:,5:15]
dropdata
```

# Aprori Algorithm

```python
frequent_itemsets = apriori(dropdata, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.7)
rules.head(11)
```

```python
rules.sort_values('lift',ascending = False).head(10)
```

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
rules[rules.lift>1]
```
